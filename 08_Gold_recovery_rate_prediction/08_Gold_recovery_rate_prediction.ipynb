### Проверим, что эффективность обогащения рассчитана правильно.
[&#10224;](#step1)

 __Вычислим эффективность обогощения на обучающей выборке для признака `rougher_output_recovery`. Найдем `MAE` между нашими расчётами и значением признака.__

```python
recovery = train[['rougher.output.concentrate_au', 'rougher.input.feed_au' , 'rougher.output.tail_au','rougher.output.recovery']]
# для наглядности, соберем исследуемые значения в одну таблицу
```

```python
# составим формулу вычисления recovery
recovery['rougher.output.recovery_check'] =  (train['rougher.output.concentrate_au'] \
                            * (train['rougher.input.feed_au'] - train['rougher.output.tail_au']))\
                            / (   train['rougher.input.feed_au'] \
                            * (train['rougher.output.concentrate_au'] - train['rougher.output.tail_au']))*100
rougher_recovery_mae = mean_absolute_error(train['rougher.output.recovery'], recovery['rougher.output.recovery_check'])
recovery['rougher_recovery_mae'] = rougher_recovery_mae
print(f'MAE между нашим расчетом и значением recovery в признаке составляет {rougher_recovery_mae:0.0%}')
recovery.head(3)
```

__Построим гистограммы распределения для этапов очистки__

```python
#Создадим функцию для автоматизации построения гистограмм
def histogram_show(data,rougher_input_feed, final_concentration, primary_concentration, rougher_concentration,t):
    data[final_concentration].hist(bins=100,figsize=(12, 7))#финальная очистка
           
    data[primary_concentration].hist(bins=100,figsize=(12, 7))# этап первичная очистка
    
    data[rougher_concentration].hist(bins=100,figsize=(12, 7))# этап фолотация выход концентрата
    data[rougher_input_feed].hist(bins=100,figsize=(12, 7))# этап фолотация входная подача
    plt.legend([final_concentration, primary_concentration,rougher_concentration, rougher_input_feed])
     
    if  t == 'концентрация металлов':
        plt.title('Гистограммы распределения концентрации металла на  этапе очистки')
    if t ==  'сумма концентрации':
        plt.title('Гистограммы распределения суммарной концентрации всех веществ на технологических стадиях')
    
    plt.ylabel('частота концентрации металла на этапах очистки')
    plt.grid(False)
    plt.show()   
```

__Создадим функцию `remove_outlier`, в которой сможем убирать значения концентраций металлов на различных стадиях очистки из датасетов.  Т.е значения не попадающие в интересующий нас интервал, и не актуальных  для исследования__

```python
def remove_outlier(d,num):     
    f = d.loc[(d['rougher.input.feed_ag'] > num) &(d['rougher.input.feed_pb'] > num)\
          & (d['rougher.input.feed_au'] > num) & (d['rougher.output.concentrate_ag'] > num)\
          & (d['rougher.output.concentrate_pb'] > num) &(d['rougher.output.concentrate_au'] > num)\
          & (d['primary_cleaner.output.concentrate_ag'] > num) & (d['primary_cleaner.output.concentrate_pb'] > num)\
          & (d['primary_cleaner.output.concentrate_au'] > num) &(d['final.output.concentrate_ag'] > num)\
          & (d['final.output.concentrate_pb'] > num)  & (d['final.output.concentrate_au'] > num)]   
    print('Изначальный размер таблицы',d.shape)
    print('Размер после удаления выбросов',f.shape)
    return f
```

__Суммируем концентрацию всех веществ на технологических стадиях получения золота__

```python
# сумма всех веществ на этапе  фолотация входная подача
full['rougher_input_sum'] = full['rougher.input.feed_ag'] + full['rougher.input.feed_pb'] \
                         + full['rougher.input.feed_au'] + full['rougher.input.feed_sol' ]
# сумма всех веществ на этапе   выходная подача черновой концентрат
full['rougher_output_sum'] = full['rougher.output.concentrate_ag'] + full['rougher.output.concentrate_pb'] \
                         + full['rougher.output.concentrate_au'] + full['rougher.output.concentrate_sol']
# сумма всех веществ на этапе  первичная очистка 
full['primary_cleaner_sum'] = full['primary_cleaner.output.concentrate_ag'] + full['primary_cleaner.output.concentrate_pb']\
                         + full['primary_cleaner.output.concentrate_au'] + full['primary_cleaner.output.concentrate_sol']
# сумма всех веществ на этапе  финальная очистка
full['final_output_sum'] = full['final.output.concentrate_ag'] + full['final.output.concentrate_pb'] \
                         + full['final.output.concentrate_au']  + full['final.output.concentrate_sol']
```

__Создадим функцию для нахождения sMAPE «симметричное среднее абсолютное процентное отклонение»).__

```python
# функцию для расчета метрики sMAPE
def sMAPE(target, predict):
    return 1/len(target) * np.sum(2 * np.abs(target - predict) / (np.abs(target) + np.abs(predict))*100)
```

<font color='green'> `smape` расчитана верно.
